```python
# scipy.spatial can compute triangulations, Voronoi diagrams an convex hulls
import scipy as sp
from scipy.spatial import Delaunay
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[0,0],[0,1.1],[1,0],[1,1]])
tri = Delaunay(points)
print(tri)
```

```python
# all method calls associated with tri
print(dir(tri))
```

```python
plt.style.use('ggplot')
plt.triplot(points[:,0], points[:,1], tri.simplices, color = 'k')
plt.plot(points[:,0], points[:,1], 'o', color = 'r')
plt.xlim(-0.5, 1.5); plt.ylim(-0.5, 1.5)
plt.show()
```

```python
# co-ordinates of the data points
# the rows are enumerated 0,1,2,3 and used to represent 0 simplices
tri.points
```

```python
tri.simplices
```

```python
plt.triplot(points[:,0], points[:,1], tri.simplices, color = 'k')
plt.plot(points[:,0], points[:,1], 'o', color = 'r')
for j, p in enumerate(points):
    plt.text(p[0]-0.03, p[1]+0.03, j, ha='right') # label the points

for j, s in enumerate(tri.simplices):
    p = points[s].mean(axis=0)
    plt.text(p[0], p[1], '#%d' % j, ha='center') # label triangles

plt.xlim(-0.5, 1.5); plt.ylim(-0.5, 1.5)
plt.show()
```
